# WEB-usb

Los navegadores Crome,Edgue pueden acceder a los dispositicos USB-serie.

Gracias a la API WebUSB que proporciona una forma segura de exponer los servicios de dispositivos USB a la web, permitiendo el acceso a dispositivos USB no est√°ndar sin necesidad de controladores propietarios o SDK nativos.

Hay varias url para acceder online. (busca: console serial web usb).   
pero no todas con compatibles con todos los dispositivos.


- Un ejemplo seria esta url:   
  https://googlechromelabs.github.io/serial-terminal/

___


1) Accede a https://googlechromelabs.github.io/serial-terminal/   
   Desde un navegador compatible ( chrome / edgue / electron .... ) 


2) Pulsa conectar y selecciona el puerto de la lista de dispositivos.   
   Nota: si no aparece es posible que la web no sea compatible. (*** VID:PID)


3) ya tienes acceso a la terminal.   
   Si pulsas intro o Ctrl+D veras el interprete Micro/Circuit-Python
